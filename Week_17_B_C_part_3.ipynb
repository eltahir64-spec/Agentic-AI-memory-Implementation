##Cell 2: The Full AI Genesis Hub (Integrated Logic & Theory)
This cell creates the app.py file. It follows the RDI Framework and uses Tier 1–4 Streamlit capabilities to ensure a professional-grade output. This version includes the Stable Table (Tier 1) alongside the Plotly Graphs (Tier 4) for maximum reliability.

```python
%%writefile app.py
import streamlit as st
from crewai import Agent, Task, Crew
import pandas as pd
import plotly.express as px

# ---- STAGE 1 & 2: THE HULL ----
st.set_page_config(page_title="AI App Genesis Hub", layout="wide")

with st.sidebar:
    st.title("⚙️ Mission Control")
    st.info("Week 17 Framework: V-Model Implementation")
    viz_style = st.selectbox("Intelligence View", ["Bar Chart", "Pie Chart", "Scatter Plot"])
    st.divider()
    st.write("**Project Goal:** Business-Consumable Product")

# ---- STAGE 3: THE DATA GATE (STATEFUL INTERACTION) ----
if 'intel_report' not in st.session_state:
    st.session_state.intel_report = None
if 'processed_df' not in st.session_state:
    st.session_state.processed_df = None

st.title('🚀 AI App Genesis: Data Intelligence Hub')

# ---- STAGE 4: REAL-TIME RISK SENSORS ----
if st.session_state.processed_df is not None:
    avg_risk = st.session_state.processed_df['Risk_Score'].mean()
    if avg_risk > 0.6:
        st.error(f"⚠️ HIGH SYSTEM RISK DETECTED: {avg_risk:.2f}")
    else:
        st.success(f"✅ System Stability Nominal: {avg_risk:.2f}")

col1, col2 = st.columns([1, 1])

with col1:
    with st.form("data_ingestion_form"):
        st.subheader("Requirement Input")
        query = st.text_input("Analysis Goal:", placeholder="e.g., Identify security risks")
        uploaded_file = st.file_uploader("Upload 'test_logs.csv'", type=['csv'])
        submitted = st.form_submit_button("Ignite Agents")

# ---- STAGE 5 & 6: THE NERVOUS SYSTEM (HANDSHAKE) ----
if submitted:
    if not query or not uploaded_file:
        st.error("Missing Requirement.")
    else:
        with st.spinner('Agents are parsing logs...'):
            df = pd.read_csv(uploaded_file)
            st.session_state.processed_df = df

            # Deep Role Engineering
            analyst = Agent(
                role="Senior AIOps Analyst",
                goal=f"Analyze logs for {query}",
                backstory="Expert in log synthesis.",
                verbose=True
            )
            task = Task(
                description=f"Objective: {query}. Data Context: Avg Risk {df['Risk_Score'].mean():.2f}.",
                agent=analyst,
                expected_output="A 3-point technical executive summary."
            )
            crew = Crew(agents=[analyst], tasks=[task])
            st.session_state.intel_report = str(crew.kickoff())

# ---- STAGE 7: RENDERING INTELLIGENCE ----
with col2:
    st.subheader("Visual Intelligence (Tier 4)")
    if st.session_state.processed_df is not None:
        df = st.session_state.processed_df

        # Adding a Stable Table for Tunnel Resilience
        st.write("### Statistical Snapshot")
        summary_table = df.groupby('Event_Type')['Traffic_Volume_MB'].sum().reset_index()
        st.table(summary_table)

        if viz_style == "Bar Chart":
            fig = px.bar(summary_table, x='Event_Type', y='Traffic_Volume_MB', template="plotly_dark")
        elif viz_style == "Pie Chart":
            fig = px.pie(df, names='Node_ID', values='Risk_Score', template="plotly_dark")
        else:
            fig = px.scatter(df, x='Traffic_Volume_MB', y='Risk_Score', color='Event_Type', template="plotly_dark")
        st.plotly_chart(fig, use_container_width=True)
    else:
        st.info("Awaiting Data Ingestion...")

if st.session_state.intel_report:
    st.divider()
    st.subheader("📜 Executive Intelligence Report")
    st.markdown(st.session_state.intel_report)
    st.download_button("Download Assessment", st.session_state.intel_report, "report.txt")
```
