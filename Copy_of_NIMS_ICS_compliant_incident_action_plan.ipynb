### Essential Components of a NIMS/ICS Incident Action Plan (IAP)

Based on NIMS/ICS guidelines (e.g., ICS Form 201, 202, 203, 204, 205, 206), a compliant Incident Action Plan (IAP) typically includes the following core sections and essential information:

**1. ICS Form 202 - Incident Objectives:**
*   **Incident Name:** Official name of the incident.
*   **Operational Period:** Start and end date/time for which the IAP is valid.
*   **General Objectives:** Overarching incident goals (safety, life safety, incident stabilization, property/environmental conservation).
*   **Specific Objectives:** Measurable, actionable, time-sensitive objectives for the operational period.
*   **Approval/Preparer Information:** Name and position of preparer, date/time prepared.

**2. ICS Form 203 - Organization Assignment List (OAL):**
*   **Incident Management Team (IMT) Structure:** Names and assignments for key ICS positions (Incident Commander, Command Staff, General Staff).
*   **Reporting Relationships:** Clear lines of authority and communication.

**3. ICS Form 204 - Assignment List:**
*   **Branch/Division/Group:** Specific operational areas.
*   **Leader:** Name of the assigned leader for the operational area.
*   **Resources Assigned:** List of personnel and equipment assigned.
*   **Work Assignments:** Specific tasks and tactics for the operational period.
*   **Special Instructions:** Any additional guidance or safety considerations.
*   **Reporting Location/Time:** Where and when resources should report.

**4. ICS Form 205 - Incident Radio Communications Plan:**
*   **Operational Period:** Dates and times.
*   **Radio System Information:** Type of system (e.g., analog, digital).
*   **Channel Assignments:** Primary, tactical, support, air-to-ground frequencies.
*   **Call Signs:** Assigned identifiers for various units.
*   **Remarks:** Any special communication procedures.

**5. ICS Form 206 - Medical Plan:**
*   **Operational Period:** Dates and times.
*   **Medical Aid Stations:** Locations and capabilities.
*   **Ambulance Exchange Points:** Designated points for patient transfer.
*   **Hospitals/Clinics:** Names, locations, contact info, and specializations.
*   **Medical Emergency Procedures:** Steps to take in case of an injury.

**6. Other Supporting Documents/Information (often attached or integrated):**
*   **Incident Map:** Depicting incident area, resources, hazards, and tactical assignments.
*   **Safety Plan (ICS 208):** Hazard analysis, safety message, control measures.
*   **Traffic Plan (if applicable):** Traffic flow, control points, access routes.
*   **Site Safety and Health Plan (SSHP):** Detailed safety procedures for hazardous materials incidents.
*   **Resource List (ICS 209/211):** Summary of all assigned resources.
*   **Weather Forecast:** Relevant weather information for the operational period.

This summary provides the foundational elements to build a comprehensive data model for an IAP, ensuring all critical information is collected and presented according to NIMS/ICS standards.

## Develop Interactive Input Script

### Subtask:
Create a Python script that interactively prompts the user for all the required information for the Incident Action Plan, following the defined data model. This script should be user-friendly and guide the user through the data entry process.


**Reasoning**:
The subtask requires creating a Python script that interactively prompts the user for IAP information based on the `iap_data_model`. I will start by defining helper functions to handle different types of inputs (strings, dates/times, lists of items, and nested dictionaries) to make the main data collection function more modular and readable. This approach aligns with the instructions to create a user-friendly and guided data entry process.



```python
import json
import datetime

def get_string_input(prompt, default=None):
    """Gets a string input from the user."""
    while True:
        value = input(prompt).strip()
        if value or default is not None:
            return value if value else default
        print("Input cannot be empty. Please provide a value.")

def get_datetime_input(prompt, format_str, value_type="date"):
    """Gets a date or time input from the user with validation."""
    while True:
        user_input = input(prompt).strip()
        if not user_input: # Allow empty input for optional fields
            return None
        try:
            if value_type == "date":
                return datetime.datetime.strptime(user_input, format_str).strftime('%Y-%m-%d')
            elif value_type == "time":
                return datetime.datetime.strptime(user_input, format_str).strftime('%H:%M')
            else:
                raise ValueError("Invalid value_type for get_datetime_input")
        except ValueError:
            print(f"Invalid format. Please use {format_str}.")

def get_list_of_dicts_input(prompt_singular, field_definitions):
    """Collects multiple entries, each a dictionary, for a list type field."""
    items = []
    print(f"\n--- Collecting {prompt_singular}s ---")
    while True:
        print(f"Enter details for a new {prompt_singular} (or type 'done' to finish):")
        temp_dict = {}
        is_done = False

        # Check if user wants to stop before collecting any fields for the current item
        check_done = input(f"    (Type 'done' or press Enter to add a {prompt_singular}): ").strip().lower()
        if check_done == 'done':
            break

        for field_name, field_type_info in field_definitions.items():
            field_type = field_type_info.get('type')
            field_prompt = field_type_info.get('prompt', f"    Enter {prompt_singular} {field_name}: ")
            field_format = field_type_info.get('format_str')

            if field_type == 'string':
                value = get_string_input(field_prompt)
            elif field_type == 'date':
                value = get_datetime_input(field_prompt, field_format, value_type='date')
            elif field_type == 'time':
                value = get_datetime_input(field_prompt, field_format, value_type='time')
            else: # Default to string for unknown types
                value = get_string_input(field_prompt)

            temp_dict[field_name] = value

        if temp_dict: # Only add if at least some fields were filled (user didn't just type 'done' and then nothing)
            items.append(temp_dict)

        # Offer to continue or stop
        if not get_string_input(f"Add another {prompt_singular}? (yes/no): ", default='yes').lower() == 'yes':
            break
    return items

def get_dict_input(prompt, field_definitions):
    """Collects fields for a dictionary type field."""
    print(f"\n--- Collecting {prompt} Details ---")
    data_dict = {}
    for field_name, field_type_info in field_definitions.items():
        field_type = field_type_info.get('type')
        field_prompt = field_type_info.get('prompt', f"    Enter {field_name}: ")
        field_format = field_type_info.get('format_str')

        if field_type == 'string':
            value = get_string_input(field_prompt)
        elif field_type == 'date':
            value = get_datetime_input(field_prompt, field_format, value_type='date')
        elif field_type == 'time':
            value = get_datetime_input(field_prompt, field_format, value_type='time')
        else:
            value = get_string_input(field_prompt)
        data_dict[field_name] = value
    return data_dict

print("Helper functions for interactive input are defined.")
```

**Reasoning**:
Now that the helper functions are defined, I will implement the main `collect_iap_data` function. This function will iterate through the `iap_data_model` structure, using the previously defined helper functions to interactively prompt the user for all necessary information. It will handle various field types including simple strings, dates, times, nested dictionaries, and lists of dictionaries. Finally, it will call this function and print the collected data to verify its structure.



```python
def collect_iap_data(data_model):
    """Interactively collects Incident Action Plan data from the user."""
    iap_data = data_model.copy() # Create a mutable copy to populate

    print("\n--- Incident Action Plan Data Collection ---")

    # Incident Name
    iap_data["Incident_Name"] = get_string_input("Enter Incident Name: ")

    # Operational Period
    print("\n--- Operational Period ---")
    iap_data["Operational_Period"]["Start_Date"] = get_datetime_input("Enter Operational Period Start Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["End_Date"] = get_datetime_input("Enter Operational Period End Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["Start_Time"] = get_datetime_input("Enter Operational Period Start Time (HH:MM): ", "%H:%M", "time")
    iap_data["Operational_Period"]["End_Time"] = get_datetime_input("Enter Operational Period End Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 202 - Incident Objectives
    print("\n--- ICS Form 202 - Incident Objectives ---")
    # Objectives - List of strings for simplicity, could be enhanced to list of dicts if more detail needed per obj
    iap_data["ICS_202_Incident_Objectives"]["Objectives"] = []
    print("Enter General Objectives (type 'done' to finish each objective):")
    while True:
        objective = get_string_input("    Objective: ")
        if objective.lower() == 'done':
            break
        if objective: # Only add if not empty
            iap_data["ICS_202_Incident_Objectives"]["Objectives"].append(objective)

    iap_data["ICS_202_Incident_Objectives"]["General_Safety_Message"] = get_string_input("Enter General Safety Message: ")

    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Name"] = get_string_input("Enter Preparer Name: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Position"] = get_string_input("Enter Preparer Position: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Signature"] = get_string_input("Enter Preparer Signature (or leave blank): ", default='')
    iap_data["ICS_202_Incident_Objectives"]["Date"] = get_datetime_input("Enter Prepared Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["ICS_202_Incident_Objectives"]["Time"] = get_datetime_input("Enter Prepared Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 203 - Organization Assignment List
    print("\n--- ICS Form 203 - Organization Assignment List ---")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Name"] = get_string_input("Enter Incident Commander Name: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Agency"] = get_string_input("Enter Incident Commander Agency: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Phone"] = get_string_input("Enter Incident Commander Phone: ")

    for section, details in iap_data["ICS_203_Organization_Assignment_List"]["Section_Chiefs"].items():
        print(f"  --- {section} Section Chief ---")
        details["Name"] = get_string_input(f"    Enter {section} Section Chief Name: ")
        details["Agency"] = get_string_input(f"    Enter {section} Section Chief Agency: ")
        details["Phone"] = get_string_input(f"    Enter {section} Section Chief Phone: ")

    # Other Positions (list of dicts)
    iap_data["ICS_203_Organization_Assignment_List"]["Other_Positions"] = get_list_of_dicts_input(
        "Other Position",
        {
            "Name": {"type": "string", "prompt": "    Enter Name: "},
            "Position": {"type": "string", "prompt": "    Enter Position: "},
            "Agency": {"type": "string", "prompt": "    Enter Agency: "},
            "Phone": {"type": "string", "prompt": "    Enter Phone: "}
        }
    )

    # ICS Form 204 - Assignment List
    iap_data["ICS_204_Assignment_List"] = get_list_of_dicts_input(
        "Assignment",
        {
            "Division_Group": {"type": "string"},
            "Branch": {"type": "string", "prompt": "    Enter Branch (optional, press Enter to skip): ", "default": None},
            "Resources_Assigned": {"type": "list_of_strings", "prompt": "    Enter Resource (e.g., Engine 1, Medic 5 - type 'done' to finish resources): "},
            "Task_Activity": {"type": "string"},
            "Special_Instructions": {"type": "string", "prompt": "    Enter Special Instructions (optional, press Enter to skip): ", "default": None},
            "Reporting_Location": {"type": "string"},
            "Reporting_Time": {"type": "time", "format_str": "%H:%M"},
            "Supervisor_Name": {"type": "string", "prompt": "    Enter Supervisor Name: "},
            "Supervisor_Contact": {"type": "string", "prompt": "    Enter Supervisor Contact: "}
        }
    )
    # Special handling for Resources_Assigned within Assignment List
    for assignment in iap_data["ICS_204_Assignment_List"]:
        resources = []
        print(f"  --- Resources for {assignment.get('Division_Group', 'an assignment')} ---")
        while True:
            resource = get_string_input("    Enter Resource (e.g., Engine 1, Medic 5 - type 'done' to finish): ")
            if resource.lower() == 'done':
                break
            if resource:
                resources.append(resource)
        assignment["Resources_Assigned"] = resources
        # Flatting supervisor details
        assignment["Supervisor"] = {
            "Name": assignment.pop("Supervisor_Name"),
            "Contact": assignment.pop("Supervisor_Contact")
        }

    # ICS Form 205 - Incident Radio Communications Plan
    iap_data["ICS_205_Incident_Radio_Communications_Plan"] = get_list_of_dicts_input(
        "Radio Channel",
        {
            "Channel_Name": {"type": "string"},
            "Function": {"type": "string"},
            "Assignment": {"type": "string"},
            "Frequency_Mode_Tone": {"type": "string"},
            "Remarks": {"type": "string", "prompt": "    Enter Remarks (optional, press Enter to skip): ", "default": None}
        }
    )

    # ICS Form 206 - Medical Plan
    print("\n--- ICS Form 206 - Medical Plan ---")
    iap_data["ICS_206_Medical_Plan"]["Medical_Aid_Stations"] = get_list_of_dicts_input(
        "Medical Aid Station",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Capabilities": {"type": "string"}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Hospitals"] = get_list_of_dicts_input(
        "Hospital",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Contact": {"type": "string"},
            "Specializations": {"type": "string", "prompt": "    Enter Specializations (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Ambulance_Services"] = get_list_of_dicts_input(
        "Ambulance Service",
        {
            "Provider": {"type": "string"},
            "Contact": {"type": "string"},
            "Exchange_Points": {"type": "string", "prompt": "    Enter Exchange Points (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Medical_Emergency_Procedures"] = get_string_input("Enter Medical Emergency Procedures: ")

    # Map/Diagram
    print("\n--- Map / Diagram ---")
    iap_data["Map_Diagram"]["Description"] = get_string_input("Enter Map/Diagram Description (optional, press Enter to skip): ", default=None)
    iap_data["Map_Diagram"]["File_Path"] = get_string_input("Enter Map/Diagram File Path/URL (optional, press Enter to skip): ", default=None)

    # Safety Message/Plan
    iap_data["Safety_Message_Plan"] = get_string_input("Enter Safety Message/Plan details (optional, press Enter to skip): ", default=None)

    return iap_data

# Call the function to collect data
print("Starting interactive IAP data collection. Follow the prompts.")
filled_iap_data = collect_iap_data(iap_data_model)

print("\n--- Collected IAP Data ---")
print(json.dumps(filled_iap_data, indent=4))
```

## Generate Formatted IAP Output

### Subtask:
Implement the Python code to take the data collected from the user and format it into a clear, readable, and NIMS/ICS compliant Incident Action Plan document. The output should be easily consumable, perhaps as a plain text or Markdown file.


**Reasoning**:
The subtask requires generating a formatted IAP document from the collected data. I will define a Python function `generate_iap_document` that takes the `filled_iap_data` and formats it into a Markdown string, handling various data types and nested structures, and then print the result.



```python
def generate_iap_document(iap_data):
    """Generates a formatted NIMS/ICS compliant Incident Action Plan document from collected data."""
    document_parts = []

    document_parts.append("# Incident Action Plan (IAP)\n")

    if iap_data.get("Incident_Name"):
        document_parts.append(f"## Incident Name: {iap_data['Incident_Name']}\n")

    # Operational Period
    op_period = iap_data.get("Operational_Period")
    if op_period and op_period.get("Start_Date"):
        document_parts.append("### 1. Operational Period\n")
        document_parts.append(f"* **Start:** {op_period.get('Start_Date')} {op_period.get('Start_Time')}\n")
        document_parts.append(f"* **End:** {op_period.get('End_Date')} {op_period.get('End_Time')}\n\n")

    # ICS 202 - Incident Objectives
    ics_202 = iap_data.get("ICS_202_Incident_Objectives")
    if ics_202 and (ics_202.get("Objectives") or ics_202.get("General_Safety_Message")):
        document_parts.append("### 2. ICS Form 202 - Incident Objectives\n")
        if ics_202.get("Objectives"):
            document_parts.append("#### Objectives:\n")
            for obj in ics_202["Objectives"]:
                document_parts.append(f"* {obj}\n")
        if ics_202.get("General_Safety_Message"):
            document_parts.append(f"#### General Safety Message:\n{ics_202['General_Safety_Message']}\n")
        if ics_202.get("Prepared_By") and ics_202["Prepared_By"].get("Name"):
            prep = ics_202["Prepared_By"]
            document_parts.append(f"* **Prepared By:** {prep.get('Name')} ({prep.get('Position')})\n")
            if prep.get('Signature'):
                document_parts.append(f"* **Signature:** {prep.get('Signature')}\n")
            document_parts.append(f"* **Date/Time:** {ics_202.get('Date')} {ics_202.get('Time')}\n\n")

    # ICS 203 - Organization Assignment List
    ics_203 = iap_data.get("ICS_203_Organization_Assignment_List")
    if ics_203 and (ics_203.get("Incident_Commander") or ics_203.get("Section_Chiefs") or ics_203.get("Other_Positions")):
        document_parts.append("### 3. ICS Form 203 - Organization Assignment List\n")
        if ics_203.get("Incident_Commander") and ics_203["Incident_Commander"].get("Name"):
            ic = ics_203["Incident_Commander"]
            document_parts.append(f"#### Incident Commander:\n")
            document_parts.append(f"* **Name:** {ic.get('Name')}\n")
            document_parts.append(f"* **Agency:** {ic.get('Agency')}\n")
            document_parts.append(f"* **Phone:** {ic.get('Phone')}\n\n")

        if ics_203.get("Section_Chiefs"):
            document_parts.append("#### Section Chiefs:\n")
            for section, details in ics_203["Section_Chiefs"].items():
                if details.get("Name"):
                    document_parts.append(f"* **{section.replace('_', ' ')}:** {details.get('Name')} ({details.get('Agency')}) - {details.get('Phone')}\n")
            document_parts.append("\n")

        if ics_203.get("Other_Positions"):
            document_parts.append("#### Other Key Positions:\n")
            for pos in ics_203["Other_Positions"]:
                if pos.get("Name"):
                    document_parts.append(f"* **{pos.get('Position')}:** {pos.get('Name')} ({pos.get('Agency')}) - {pos.get('Phone')}\n")
            document_parts.append("\n")

    # ICS 204 - Assignment List
    ics_204_list = iap_data.get("ICS_204_Assignment_List")
    if ics_204_list:
        document_parts.append("### 4. ICS Form 204 - Assignment List\n")
        for i, assignment in enumerate(ics_204_list):
            if assignment.get("Division_Group"):
                document_parts.append(f"#### Assignment {i+1}: {assignment.get('Division_Group')}\n")
                if assignment.get("Branch"):
                    document_parts.append(f"* **Branch:** {assignment.get('Branch')}\n")
                if assignment.get("Resources_Assigned"):
                    document_parts.append(f"* **Resources Assigned:** {', '.join(assignment['Resources_Assigned'])}\n")
                if assignment.get("Task_Activity"):
                    document_parts.append(f"* **Task/Activity:** {assignment.get('Task_Activity')}\n")
                if assignment.get("Special_Instructions"):
                    document_parts.append(f"* **Special Instructions:** {assignment.get('Special_Instructions')}\n")
                if assignment.get("Reporting_Location"):
                    document_parts.append(f"* **Reporting Location:** {assignment.get('Reporting_Location')}\n")
                if assignment.get("Reporting_Time"):
                    document_parts.append(f"* **Reporting Time:** {assignment.get('Reporting_Time')}\n")
                if assignment.get("Supervisor") and assignment["Supervisor"].get("Name"):
                    sup = assignment["Supervisor"]
                    document_parts.append(f"* **Supervisor:** {sup.get('Name')} ({sup.get('Contact')})\n")
                document_parts.append("\n")

    # ICS 205 - Incident Radio Communications Plan
    ics_205_list = iap_data.get("ICS_205_Incident_Radio_Communications_Plan")
    if ics_205_list:
        document_parts.append("### 5. ICS Form 205 - Incident Radio Communications Plan\n")
        for i, channel in enumerate(ics_205_list):
            if channel.get("Channel_Name"):
                document_parts.append(f"#### Channel {i+1}: {channel.get('Channel_Name')}\n")
                document_parts.append(f"* **Function:** {channel.get('Function')}\n")
                document_parts.append(f"* **Assignment:** {channel.get('Assignment')}\n")
                document_parts.append(f"* **Frequency/Mode/Tone:** {channel.get('Frequency_Mode_Tone')}\n")
                if channel.get("Remarks"):
                    document_parts.append(f"* **Remarks:** {channel.get('Remarks')}\n")
                document_parts.append("\n")

    # ICS 206 - Medical Plan
    ics_206 = iap_data.get("ICS_206_Medical_Plan")
    if ics_206 and (ics_206.get("Medical_Aid_Stations") or ics_206.get("Hospitals") or ics_206.get("Ambulance_Services") or ics_206.get("Medical_Emergency_Procedures")):
        document_parts.append("### 6. ICS Form 206 - Medical Plan\n")
        if ics_206.get("Medical_Aid_Stations"):
            document_parts.append("#### Medical Aid Stations:\n")
            for station in ics_206["Medical_Aid_Stations"]:
                if station.get("Name"):
                    document_parts.append(f"* **{station.get('Name')}:** {station.get('Location')} (Capabilities: {station.get('Capabilities')})\n")
            document_parts.append("\n")

        if ics_206.get("Hospitals"):
            document_parts.append("#### Hospitals:\n")
            for hospital in ics_206["Hospitals"]:
                if hospital.get("Name"):
                    specs = f" (Specializations: {hospital['Specializations']})" if hospital.get('Specializations') else ''
                    document_parts.append(f"* **{hospital.get('Name')}:** {hospital.get('Location')} (Contact: {hospital.get('Contact')}){specs}\n")
            document_parts.append("\n")

        if ics_206.get("Ambulance_Services"):
            document_parts.append("#### Ambulance Services:\n")
            for service in ics_206["Ambulance_Services"]:
                if service.get("Provider"):
                    ex_points = f" (Exchange Points: {service['Exchange_Points']})" if service.get('Exchange_Points') else ''
                    document_parts.append(f"* **{service.get('Provider')}:** Contact: {service.get('Contact')}{ex_points}\n")
            document_parts.append("\n")

        if ics_206.get("Medical_Emergency_Procedures"):
            document_parts.append(f"#### Medical Emergency Procedures:\n{ics_206['Medical_Emergency_Procedures']}\n\n")

    # Map/Diagram
    map_diag = iap_data.get("Map_Diagram")
    if map_diag and (map_diag.get("Description") or map_diag.get("File_Path")):
        document_parts.append("### 7. Map / Diagram\n")
        if map_diag.get("Description"):
            document_parts.append(f"* **Description:** {map_diag.get('Description')}\n")
        if map_diag.get("File_Path"):
            document_parts.append(f"* **File Path/URL:** {map_diag.get('File_Path')}\n")
        document_parts.append("\n")

    # Safety Message/Plan
    if iap_data.get("Safety_Message_Plan"):
        document_parts.append("### 8. Safety Message / Plan\n")
        document_parts.append(f"{iap_data['Safety_Message_Plan']}\n\n")

    return "".join(document_parts)

# Generate and print the IAP document
print("\n--- Generated IAP Document ---\n")
formatted_iap_document = generate_iap_document(filled_iap_data)
print(formatted_iap_document)
```

```python
import json
import datetime

def get_string_input(prompt, default=None):
    """Gets a string input from the user."""
    while True:
        value = input(prompt).strip()
        if value or default is not None:
            return value if value else default
        print("Input cannot be empty. Please provide a value.")

def get_datetime_input(prompt, format_str, value_type="date"):
    """Gets a date or time input from the user with validation."""
    while True:
        user_input = input(prompt).strip()
        if not user_input: # Allow empty input for optional fields
            return None
        try:
            if value_type == "date":
                return datetime.datetime.strptime(user_input, format_str).strftime('%Y-%m-%d')
            elif value_type == "time":
                return datetime.datetime.strptime(user_input, format_str).strftime('%H:%M')
            else:
                raise ValueError("Invalid value_type for get_datetime_input")
        except ValueError:
            print(f"Invalid format. Please use {format_str}.")

def get_list_of_dicts_input(prompt_singular, field_definitions):
    """Collects multiple entries, each a dictionary, for a list type field."""
    items = []
    print(f"\n--- Collecting {prompt_singular}s ---")
    while True:
        print(f"Enter details for a new {prompt_singular} (or type 'done' to finish):")
        temp_dict = {}

        # Check if user wants to stop before collecting any fields for the current item
        check_done = input(f"    (Type 'done' or press Enter to add a {prompt_singular}): ").strip().lower()
        if check_done == 'done':
            break

        all_fields_empty = True
        for field_name, field_type_info in field_definitions.items():
            field_type = field_type_info.get('type')
            field_prompt = field_type_info.get('prompt', f"    Enter {prompt_singular} {field_name}: ")
            field_format = field_type_info.get('format_str')
            field_default = field_type_info.get('default')

            if field_type == 'string':
                value = get_string_input(field_prompt, default=field_default)
            elif field_type == 'date':
                value = get_datetime_input(field_prompt, field_format, value_type='date')
            elif field_type == 'time':
                value = get_datetime_input(field_prompt, field_format, value_type='time')
            else: # Default to string for unknown types
                value = get_string_input(field_prompt, default=field_default)

            temp_dict[field_name] = value
            if value is not None and value != '': # Check if any field has a non-empty value
                all_fields_empty = False

        if temp_dict and not all_fields_empty: # Only add if at least some fields were filled
            items.append(temp_dict)

        # Offer to continue or stop
        if not get_string_input(f"Add another {prompt_singular}? (yes/no): ", default='yes').lower() == 'yes':
            break
    return items

def get_dict_input(prompt, field_definitions):
    """Collects fields for a dictionary type field."""
    print(f"\n--- Collecting {prompt} Details ---")
    data_dict = {}
    for field_name, field_type_info in field_definitions.items():
        field_type = field_type_info.get('type')
        field_prompt = field_type_info.get('prompt', f"    Enter {field_name}: ")
        field_format = field_type_info.get('format_str')
        field_default = field_type_info.get('default')

        if field_type == 'string':
            value = get_string_input(field_prompt, default=field_default)
        elif field_type == 'date':
            value = get_datetime_input(field_prompt, field_format, value_type='date')
        elif field_type == 'time':
            value = get_datetime_input(field_prompt, field_format, value_type='time')
        else:
            value = get_string_input(field_prompt, default=field_default)
        data_dict[field_name] = value
    return data_dict

# Existing IAP data model from previous steps
iap_data_model = {
    "Incident_Name": None,
    "Operational_Period": {
        "Start_Date": None,
        "End_Date": None,
        "Start_Time": None,
        "End_Time": None
    },
    "ICS_202_Incident_Objectives": {
        "Objectives": [],
        "General_Safety_Message": None,
        "Prepared_By": {
            "Name": None,
            "Position": None,
            "Signature": None
        },
        "Date": None,
        "Time": None
    },
    "ICS_203_Organization_Assignment_List": {
        "Incident_Commander": {
            "Name": None,
            "Agency": None,
            "Phone": None
        },
        "Section_Chiefs": {
            "Operations": {"Name": None, "Agency": None, "Phone": None},
            "Planning": {"Name": None, "Agency": None, "Phone": None},
            "Logistics": {"Name": None, "Agency": None, "Phone": None},
            "Finance_Admin": {"Name": None, "Agency": None, "Phone": None}
        },
        "Other_Positions": [] # List of dictionaries for other key positions
    },
    "ICS_204_Assignment_List": [
        # Each item in this list will be a dictionary representing one assignment
        {
            "Division_Group": None,
            "Branch": None,
            "Resources_Assigned": [], # List of dictionaries for resources (e.g., {'Name': 'Engine 1', 'Type': 'Fire Engine'})
            "Task_Activity": None,
            "Special_Instructions": None,
            "Reporting_Location": None,
            "Reporting_Time": None,
            "Supervisor": {"Name": None, "Contact": None}
        }
    ],
    "ICS_205_Incident_Radio_Communications_Plan": [
        # Each item in this list will be a dictionary representing one channel
        {
            "Channel_Name": None,
            "Function": None,
            "Assignment": None,
            "Frequency_Mode_Tone": None,
            "Remarks": None
        }
    ],
    "ICS_206_Medical_Plan": {
        "Medical_Aid_Stations": [],
        "Hospitals": [],
        "Ambulance_Services": [],
        "Medical_Emergency_Procedures": None
    },
    "Map_Diagram": {
        "Description": None,
        "File_Path": None # Placeholder for a file path or URL to a map/diagram
    },
    "Safety_Message_Plan": None # Could be a string or a more structured dictionary
}

def collect_iap_data(data_model):
    """Interactively collects Incident Action Plan data from the user."""
    iap_data = data_model.copy() # Create a mutable copy to populate

    print("\n--- Incident Action Plan Data Collection ---")

    # Incident Name
    iap_data["Incident_Name"] = get_string_input("Enter Incident Name: ")

    # Operational Period
    print("\n--- Operational Period ---")
    iap_data["Operational_Period"]["Start_Date"] = get_datetime_input("Enter Operational Period Start Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["End_Date"] = get_datetime_input("Enter Operational Period End Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["Start_Time"] = get_datetime_input("Enter Operational Period Start Time (HH:MM): ", "%H:%M", "time")
    iap_data["Operational_Period"]["End_Time"] = get_datetime_input("Enter Operational Period End Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 202 - Incident Objectives
    print("\n--- ICS Form 202 - Incident Objectives ---")
    # Objectives - List of strings for simplicity, could be enhanced to list of dicts if more detail needed per obj
    iap_data["ICS_202_Incident_Objectives"]["Objectives"] = []
    print("Enter General Objectives (type 'done' to finish each objective):")
    while True:
        objective = get_string_input("    Objective: ")
        if objective.lower() == 'done':
            break
        if objective: # Only add if not empty
            iap_data["ICS_202_Incident_Objectives"]["Objectives"].append(objective)

    iap_data["ICS_202_Incident_Objectives"]["General_Safety_Message"] = get_string_input("Enter General Safety Message: ")

    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Name"] = get_string_input("Enter Preparer Name: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Position"] = get_string_input("Enter Preparer Position: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Signature"] = get_string_input("Enter Preparer Signature (or leave blank): ", default='')
    iap_data["ICS_202_Incident_Objectives"]["Date"] = get_datetime_input("Enter Prepared Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["ICS_202_Incident_Objectives"]["Time"] = get_datetime_input("Enter Prepared Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 203 - Organization Assignment List
    print("\n--- ICS Form 203 - Organization Assignment List ---")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Name"] = get_string_input("Enter Incident Commander Name: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Agency"] = get_string_input("Enter Incident Commander Agency: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Phone"] = get_string_input("Enter Incident Commander Phone: ")

    for section, details in iap_data["ICS_203_Organization_Assignment_List"]["Section_Chiefs"].items():
        print(f"  --- {section} Section Chief ---")
        details["Name"] = get_string_input(f"    Enter {section} Section Chief Name: ")
        details["Agency"] = get_string_input(f"    Enter {section} Section Chief Agency: ")
        details["Phone"] = get_string_input(f"    Enter {section} Section Chief Phone: ")

    # Other Positions (list of dicts)
    iap_data["ICS_203_Organization_Assignment_List"]["Other_Positions"] = get_list_of_dicts_input(
        "Other Position",
        {
            "Name": {"type": "string", "prompt": "    Enter Name: "},
            "Position": {"type": "string", "prompt": "    Enter Position: "},
            "Agency": {"type": "string", "prompt": "    Enter Agency: "},
            "Phone": {"type": "string", "prompt": "    Enter Phone: "}
        }
    )

    # ICS Form 204 - Assignment List
    iap_data["ICS_204_Assignment_List"] = get_list_of_dicts_input(
        "Assignment",
        {
            "Division_Group": {"type": "string"},
            "Branch": {"type": "string", "prompt": "    Enter Branch (optional, press Enter to skip): ", "default": None},
            "Task_Activity": {"type": "string"},
            "Special_Instructions": {"type": "string", "prompt": "    Enter Special Instructions (optional, press Enter to skip): ", "default": None},
            "Reporting_Location": {"type": "string"},
            "Reporting_Time": {"type": "time", "format_str": "%H:%M"},
            "Supervisor_Name": {"type": "string", "prompt": "    Enter Supervisor Name: "},
            "Supervisor_Contact": {"type": "string", "prompt": "    Enter Supervisor Contact: "}
        }
    )
    # Special handling for Resources_Assigned within Assignment List
    for assignment in iap_data["ICS_204_Assignment_List"]:
        resources = []
        print(f"  --- Resources for {assignment.get('Division_Group', 'an assignment')} ---")
        while True:
            resource = get_string_input("    Enter Resource (e.g., Engine 1, Medic 5 - type 'done' to finish): ")
            if resource.lower() == 'done':
                break
            if resource:
                resources.append(resource)
        assignment["Resources_Assigned"] = resources
        # Flatting supervisor details
        assignment["Supervisor"] = {
            "Name": assignment.pop("Supervisor_Name"),
            "Contact": assignment.pop("Supervisor_Contact")
        }

    # ICS Form 205 - Incident Radio Communications Plan
    iap_data["ICS_205_Incident_Radio_Communications_Plan"] = get_list_of_dicts_input(
        "Radio Channel",
        {
            "Channel_Name": {"type": "string"},
            "Function": {"type": "string"},
            "Assignment": {"type": "string"},
            "Frequency_Mode_Tone": {"type": "string"},
            "Remarks": {"type": "string", "prompt": "    Enter Remarks (optional, press Enter to skip): ", "default": None}
        }
    )

    # ICS Form 206 - Medical Plan
    print("\n--- ICS Form 206 - Medical Plan ---")
    iap_data["ICS_206_Medical_Plan"]["Medical_Aid_Stations"] = get_list_of_dicts_input(
        "Medical Aid Station",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Capabilities": {"type": "string"}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Hospitals"] = get_list_of_dicts_input(
        "Hospital",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Contact": {"type": "string"},
            "Specializations": {"type": "string", "prompt": "    Enter Specializations (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Ambulance_Services"] = get_list_of_dicts_input(
        "Ambulance Service",
        {
            "Provider": {"type": "string"},
            "Contact": {"type": "string"},
            "Exchange_Points": {"type": "string", "prompt": "    Enter Exchange Points (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Medical_Emergency_Procedures"] = get_string_input("Enter Medical Emergency Procedures: ")

    # Map/Diagram
    print("\n--- Map / Diagram ---")
    iap_data["Map_Diagram"]["Description"] = get_string_input("Enter Map/Diagram Description (optional, press Enter to skip): ", default=None)
    iap_data["Map_Diagram"]["File_Path"] = get_string_input("Enter Map/Diagram File Path/URL (optional, press Enter to skip): ", default=None)

    # Safety Message/Plan
    iap_data["Safety_Message_Plan"] = get_string_input("Enter Safety Message/Plan details (optional, press Enter to skip): ", default=None)

    return iap_data

def generate_iap_document(iap_data):
    """Generates a formatted NIMS/ICS compliant Incident Action Plan document from collected data."""
    document_parts = []

    document_parts.append("# Incident Action Plan (IAP)\n")

    if iap_data.get("Incident_Name"):
        document_parts.append(f"## Incident Name: {iap_data['Incident_Name']}\n")

    # Operational Period
    op_period = iap_data.get("Operational_Period")
    if op_period and op_period.get("Start_Date"):
        document_parts.append("### 1. Operational Period\n")
        document_parts.append(f"* **Start:** {op_period.get('Start_Date')} {op_period.get('Start_Time')}\n")
        document_parts.append(f"* **End:** {op_period.get('End_Date')} {op_period.get('End_Time')}\n\n")

    # ICS 202 - Incident Objectives
    ics_202 = iap_data.get("ICS_202_Incident_Objectives")
    if ics_202 and (ics_202.get("Objectives") or ics_202.get("General_Safety_Message")):
        document_parts.append("### 2. ICS Form 202 - Incident Objectives\n")
        if ics_202.get("Objectives"):
            document_parts.append("#### Objectives:\n")
            for obj in ics_202["Objectives"]:
                document_parts.append(f"* {obj}\n")
        if ics_202.get("General_Safety_Message"):
            document_parts.append(f"#### General Safety Message:\n{ics_202['General_Safety_Message']}\n")
        if ics_202.get("Prepared_By") and ics_202["Prepared_By"].get("Name"):
            prep = ics_202["Prepared_By"]
            document_parts.append(f"* **Prepared By:** {prep.get('Name')} ({prep.get('Position')})\n")
            if prep.get('Signature'):
                document_parts.append(f"* **Signature:** {prep.get('Signature')}\n")
            document_parts.append(f"* **Date/Time:** {ics_202.get('Date')} {ics_202.get('Time')}\n\n")

    # ICS 203 - Organization Assignment List
    ics_203 = iap_data.get("ICS_203_Organization_Assignment_List")
    if ics_203 and (ics_203.get("Incident_Commander") or ics_203.get("Section_Chiefs") or ics_203.get("Other_Positions")):
        document_parts.append("### 3. ICS Form 203 - Organization Assignment List\n")
        if ics_203.get("Incident_Commander") and ics_203["Incident_Commander"].get("Name"):
            ic = ics_203["Incident_Commander"]
            document_parts.append(f"#### Incident Commander:\n")
            document_parts.append(f"* **Name:** {ic.get('Name')}\n")
            document_parts.append(f"* **Agency:** {ic.get('Agency')}\n")
            document_parts.append(f"* **Phone:** {ic.get('Phone')}\n\n")

        if ics_203.get("Section_Chiefs"):
            document_parts.append("#### Section Chiefs:\n")
            for section, details in ics_203["Section_Chiefs"].items():
                if details.get("Name"):
                    document_parts.append(f"* **{section.replace('_', ' ')}:** {details.get('Name')} ({details.get('Agency')}) - {details.get('Phone')}\n")
            document_parts.append("\n")

        if ics_203.get("Other_Positions"):
            document_parts.append("#### Other Key Positions:\n")
            for pos in ics_203["Other_Positions"]:
                if pos.get("Name"):
                    document_parts.append(f"* **{pos.get('Position')}:** {pos.get('Name')} ({pos.get('Agency')}) - {pos.get('Phone')}\n")
            document_parts.append("\n")

    # ICS 204 - Assignment List
    ics_204_list = iap_data.get("ICS_204_Assignment_List")
    if ics_204_list:
        document_parts.append("### 4. ICS Form 204 - Assignment List\n")
        for i, assignment in enumerate(ics_204_list):
            if assignment.get("Division_Group"):
                document_parts.append(f"#### Assignment {i+1}: {assignment.get('Division_Group')}\n")
                if assignment.get("Branch"):
                    document_parts.append(f"* **Branch:** {assignment.get('Branch')}\n")
                if assignment.get("Resources_Assigned"):
                    document_parts.append(f"* **Resources Assigned:** {', '.join(assignment['Resources_Assigned'])}\n")
                if assignment.get("Task_Activity"):
                    document_parts.append(f"* **Task/Activity:** {assignment.get('Task_Activity')}\n")
                if assignment.get("Special_Instructions"):
                    document_parts.append(f"* **Special Instructions:** {assignment.get('Special_Instructions')}\n")
                if assignment.get("Reporting_Location"):
                    document_parts.append(f"* **Reporting Location:** {assignment.get('Reporting_Location')}\n")
                if assignment.get("Reporting_Time"):
                    document_parts.append(f"* **Reporting Time:** {assignment.get('Reporting_Time')}\n")
                if assignment.get("Supervisor") and assignment["Supervisor"].get("Name"):
                    sup = assignment["Supervisor"]
                    document_parts.append(f"* **Supervisor:** {sup.get('Name')} ({sup.get('Contact')})\n")
                document_parts.append("\n")

    # ICS 205 - Incident Radio Communications Plan
    ics_205_list = iap_data.get("ICS_205_Incident_Radio_Communications_Plan")
    if ics_205_list:
        document_parts.append("### 5. ICS Form 205 - Incident Radio Communications Plan\n")
        for i, channel in enumerate(ics_205_list):
            if channel.get("Channel_Name"):
                document_parts.append(f"#### Channel {i+1}: {channel.get('Channel_Name')}\n")
                document_parts.append(f"* **Function:** {channel.get('Function')}\n")
                document_parts.append(f"* **Assignment:** {channel.get('Assignment')}\n")
                document_parts.append(f"* **Frequency/Mode/Tone:** {channel.get('Frequency_Mode_Tone')}\n")
                if channel.get("Remarks"):
                    document_parts.append(f"* **Remarks:** {channel.get('Remarks')}\n")
                document_parts.append("\n")

    # ICS 206 - Medical Plan
    ics_206 = iap_data.get("ICS_206_Medical_Plan")
    if ics_206 and (ics_206.get("Medical_Aid_Stations") or ics_206.get("Hospitals") or ics_206.get("Ambulance_Services") or ics_206.get("Medical_Emergency_Procedures")):
        document_parts.append("### 6. ICS Form 206 - Medical Plan\n")
        if ics_206.get("Medical_Aid_Stations"):
            document_parts.append("#### Medical Aid Stations:\n")
            for station in ics_206["Medical_Aid_Stations"]:
                if station.get("Name"):
                    document_parts.append(f"* **{station.get('Name')}:** {station.get('Location')} (Capabilities: {station.get('Capabilities')})\n")
            document_parts.append("\n")

        if ics_206.get("Hospitals"):
            document_parts.append("#### Hospitals:\n")
            for hospital in ics_206["Hospitals"]:
                if hospital.get("Name"):
                    specs = f" (Specializations: {hospital['Specializations']})" if hospital.get('Specializations') else ''
                    document_parts.append(f"* **{hospital.get('Name')}:** {hospital.get('Location')} (Contact: {hospital.get('Contact')}){specs}\n")
            document_parts.append("\n")

        if ics_206.get("Ambulance_Services"):
            document_parts.append("#### Ambulance Services:\n")
            for service in ics_206["Ambulance_Services"]:
                if service.get("Provider"):
                    ex_points = f" (Exchange Points: {service['Exchange_Points']})" if service.get('Exchange_Points') else ''
                    document_parts.append(f"* **{service.get('Provider')}:** Contact: {service.get('Contact')}{ex_points}\n")
            document_parts.append("\n")

        if ics_206.get("Medical_Emergency_Procedures"):
            document_parts.append(f"#### Medical Emergency Procedures:\n{ics_206['Medical_Emergency_Procedures']}\n\n")

    # Map/Diagram
    map_diag = iap_data.get("Map_Diagram")
    if map_diag and (map_diag.get("Description") or map_diag.get("File_Path")):
        document_parts.append("### 7. Map / Diagram\n")
        if map_diag.get("Description"):
            document_parts.append(f"* **Description:** {map_diag.get('Description')}\n")
        if map_diag.get("File_Path"):
            document_parts.append(f"* **File Path/URL:** {map_diag.get('File_Path')}\n")
        document_parts.append("\n")

    # Safety Message/Plan
    if iap_data.get("Safety_Message_Plan"):
        document_parts.append("### 8. Safety Message / Plan\n")
        document_parts.append(f"{iap_data['Safety_Message_Plan']}\n\n")

    return "".join(document_parts)

def save_iap_to_files(formatted_iap_document, filename_prefix="incident_iap"):
    """Saves the formatted IAP document to Markdown and plain text files."""
    markdown_filename = f"{filename_prefix}.md"
    text_filename = f"{filename_prefix}.txt"

    try:
        with open(markdown_filename, "w", encoding="utf-8") as f:
            f.write(formatted_iap_document)
        print(f"IAP successfully saved to {markdown_filename}")

        # For simplicity, saving the Markdown content directly to the text file
        # A more robust solution might involve stripping markdown or converting to pure text.
        with open(text_filename, "w", encoding="utf-8") as f:
            f.write(formatted_iap_document)
        print(f"IAP successfully saved to {text_filename}")

    except IOError as e:
        print(f"Error saving files: {e}")


# --- Main Execution Flow ---
print("Starting interactive IAP data collection. Follow the prompts.")
filled_iap_data = collect_iap_data(iap_data_model)

print("\n--- Collected IAP Data ---")
print(json.dumps(filled_iap_data, indent=4))

print("\n--- Generating IAP Document ---")
formatted_iap_document = generate_iap_document(filled_iap_data)
print(formatted_iap_document)

# Save the generated document to files
save_iap_to_files(formatted_iap_document, filename_prefix="generated_iap")
```

```python
import markdown
from weasyprint import HTML
import os

def save_iap_to_pdf(formatted_iap_document, filename_prefix="incident_iap"):
    """Converts the Markdown formatted IAP document to HTML and then saves it as a PDF file."""
    pdf_filename = f"{filename_prefix}.pdf"

    try:
        # Convert Markdown to HTML
        html_content = markdown.markdown(formatted_iap_document)

        # Add some basic styling for better PDF output
        html_styled = f"""
<html>
<head>
    <meta charset="utf-8">
    <title>{filename_prefix.replace('_', ' ').title()} IAP</title>
    <style>
        body {{ font-family: sans-serif; margin: 1in; }}
        h1 {{ color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 0.5em; }}
        h2 {{ color: #34495e; border-bottom: 1px solid #ccc; padding-bottom: 0.3em; }}
        h3 {{ color: #34495e; margin-top: 1.5em; }}
        h4 {{ color: #34495e; margin-top: 1em; }}
        ul {{ list-style-type: disc; margin-left: 1.5em; }}
        li {{ margin-bottom: 0.5em; }}
        b {{ font-weight: bold; }}
    </style>
</head>
<body>
{html_content}
</body>
</html>
"""

        # Generate PDF from HTML
        HTML(string=html_styled).write_pdf(pdf_filename)
        print(f"IAP successfully saved to {pdf_filename}")

    except Exception as e:
        print(f"Error saving PDF file: {e}")

print("PDF export function 'save_iap_to_pdf' defined.")
```

```python
import json
import datetime
import markdown
from weasyprint import HTML
import os

def get_string_input(prompt, default=None):
    """Gets a string input from the user."""
    while True:
        value = input(prompt).strip()
        if value or default is not None:
            return value if value else default
        print("Input cannot be empty. Please provide a value.")

def get_datetime_input(prompt, format_str, value_type="date"):
    """Gets a date or time input from the user with validation."""
    while True:
        user_input = input(prompt).strip()
        if not user_input: # Allow empty input for optional fields
            return None
        try:
            if value_type == "date":
                return datetime.datetime.strptime(user_input, format_str).strftime('%Y-%m-%d')
            elif value_type == "time":
                return datetime.datetime.strptime(user_input, format_str).strftime('%H:%M')
            else:
                raise ValueError("Invalid value_type for get_datetime_input")
        except ValueError:
            print(f"Invalid format. Please use {format_str}.")

def get_list_of_dicts_input(prompt_singular, field_definitions):
    """Collects multiple entries, each a dictionary, for a list type field."""
    items = []
    print(f"\n--- Collecting {prompt_singular}s ---")
    while True:
        print(f"Enter details for a new {prompt_singular} (or type 'done' to finish):")
        temp_dict = {}

        # Check if user wants to stop before collecting any fields for the current item
        check_done = input(f"    (Type 'done' or press Enter to add a {prompt_singular}): ").strip().lower()
        if check_done == 'done':
            break

        all_fields_empty = True
        for field_name, field_type_info in field_definitions.items():
            field_type = field_type_info.get('type')
            field_prompt = field_type_info.get('prompt', f"    Enter {prompt_singular} {field_name}: ")
            field_format = field_type_info.get('format_str')
            field_default = field_type_info.get('default')

            if field_type == 'string':
                value = get_string_input(field_prompt, default=field_default)
            elif field_type == 'date':
                value = get_datetime_input(field_prompt, field_format, value_type='date')
            elif field_type == 'time':
                value = get_datetime_input(field_prompt, field_format, value_type='time')
            else: # Default to string for unknown types
                value = get_string_input(field_prompt, default=field_default)

            temp_dict[field_name] = value
            if value is not None and value != '': # Check if any field has a non-empty value
                all_fields_empty = False

        if temp_dict and not all_fields_empty: # Only add if at least some fields were filled
            items.append(temp_dict)

        # Offer to continue or stop
        if not get_string_input(f"Add another {prompt_singular}? (yes/no): ", default='yes').lower() == 'yes':
            break
    return items

def get_dict_input(prompt, field_definitions):
    """Collects fields for a dictionary type field."""
    print(f"\n--- Collecting {prompt} Details ---")
    data_dict = {}
    for field_name, field_type_info in field_definitions.items():
        field_type = field_type_info.get('type')
        field_prompt = field_type_info.get('prompt', f"    Enter {field_name}: ")
        field_format = field_type_info.get('format_str')
        field_default = field_type_info.get('default')

        if field_type == 'string':
            value = get_string_input(field_prompt, default=field_default)
        elif field_type == 'date':
            value = get_datetime_input(field_prompt, field_format, value_type='date')
        elif field_type == 'time':
            value = get_datetime_input(field_prompt, field_format, value_type='time')
        else:
            value = get_string_input(field_prompt, default=field_default)
        data_dict[field_name] = value
    return data_dict

# Existing IAP data model from previous steps
iap_data_model = {
    "Incident_Name": None,
    "Operational_Period": {
        "Start_Date": None,
        "End_Date": None,
        "Start_Time": None,
        "End_Time": None
    },
    "ICS_202_Incident_Objectives": {
        "Objectives": [],
        "General_Safety_Message": None,
        "Prepared_By": {
            "Name": None,
            "Position": None,
            "Signature": None
        },
        "Date": None,
        "Time": None
    },
    "ICS_203_Organization_Assignment_List": {
        "Incident_Commander": {
            "Name": None,
            "Agency": None,
            "Phone": None
        },
        "Section_Chiefs": {
            "Operations": {"Name": None, "Agency": None, "Phone": None},
            "Planning": {"Name": None, "Agency": None, "Phone": None},
            "Logistics": {"Name": None, "Agency": None, "Phone": None},
            "Finance_Admin": {"Name": None, "Agency": None, "Phone": None}
        },
        "Other_Positions": [] # List of dictionaries for other key positions
    },
    "ICS_204_Assignment_List": [
        # Each item in this list will be a dictionary representing one assignment
        {
            "Division_Group": None,
            "Branch": None,
            "Resources_Assigned": [], # List of dictionaries for resources (e.g., {'Name': 'Engine 1', 'Type': 'Fire Engine'})
            "Task_Activity": None,
            "Special_Instructions": None,
            "Reporting_Location": None,
            "Reporting_Time": None,
            "Supervisor": {"Name": None, "Contact": None}
        }
    ],
    "ICS_205_Incident_Radio_Communications_Plan": [
        # Each item in this list will be a dictionary representing one channel
        {
            "Channel_Name": None,
            "Function": None,
            "Assignment": None,
            "Frequency_Mode_Tone": None,
            "Remarks": None
        }
    ],
    "ICS_206_Medical_Plan": {
        "Medical_Aid_Stations": [],
        "Hospitals": [],
        "Ambulance_Services": [],
        "Medical_Emergency_Procedures": None
    },
    "Map_Diagram": {
        "Description": None,
        "File_Path": None # Placeholder for a file path or URL to a map/diagram
    },
    "Safety_Message_Plan": None # Could be a string or a more structured dictionary
}

def collect_iap_data(data_model):
    """Interactively collects Incident Action Plan data from the user."""
    iap_data = data_model.copy() # Create a mutable copy to populate

    print("\n--- Incident Action Plan Data Collection ---")

    # Incident Name
    iap_data["Incident_Name"] = get_string_input("Enter Incident Name: ")

    # Operational Period
    print("\n--- Operational Period ---")
    iap_data["Operational_Period"]["Start_Date"] = get_datetime_input("Enter Operational Period Start Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["End_Date"] = get_datetime_input("Enter Operational Period End Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["Operational_Period"]["Start_Time"] = get_datetime_input("Enter Operational Period Start Time (HH:MM): ", "%H:%M", "time")
    iap_data["Operational_Period"]["End_Time"] = get_datetime_input("Enter Operational Period End Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 202 - Incident Objectives
    print("\n--- ICS Form 202 - Incident Objectives ---")
    # Objectives - List of strings for simplicity, could be enhanced to list of dicts if more detail needed per obj
    iap_data["ICS_202_Incident_Objectives"]["Objectives"] = []
    print("Enter General Objectives (type 'done' to finish each objective):")
    while True:
        objective = get_string_input("    Objective: ")
        if objective.lower() == 'done':
            break
        if objective: # Only add if not empty
            iap_data["ICS_202_Incident_Objectives"]["Objectives"].append(objective)

    iap_data["ICS_202_Incident_Objectives"]["General_Safety_Message"] = get_string_input("Enter General Safety Message: ")

    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Name"] = get_string_input("Enter Preparer Name: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Position"] = get_string_input("Enter Preparer Position: ")
    iap_data["ICS_202_Incident_Objectives"]["Prepared_By"]["Signature"] = get_string_input("Enter Preparer Signature (or leave blank): ", default='')
    iap_data["ICS_202_Incident_Objectives"]["Date"] = get_datetime_input("Enter Prepared Date (YYYY-MM-DD): ", "%Y-%m-%d", "date")
    iap_data["ICS_202_Incident_Objectives"]["Time"] = get_datetime_input("Enter Prepared Time (HH:MM): ", "%H:%M", "time")

    # ICS Form 203 - Organization Assignment List
    print("\n--- ICS Form 203 - Organization Assignment List ---")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Name"] = get_string_input("Enter Incident Commander Name: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Agency"] = get_string_input("Enter Incident Commander Agency: ")
    iap_data["ICS_203_Organization_Assignment_List"]["Incident_Commander"]["Phone"] = get_string_input("Enter Incident Commander Phone: ")

    for section, details in iap_data["ICS_203_Organization_Assignment_List"]["Section_Chiefs"].items():
        print(f"  --- {section} Section Chief ---")
        details["Name"] = get_string_input(f"    Enter {section} Section Chief Name: ")
        details["Agency"] = get_string_input(f"    Enter {section} Section Chief Agency: ")
        details["Phone"] = get_string_input(f"    Enter {section} Section Chief Phone: ")

    # Other Positions (list of dicts)
    iap_data["ICS_203_Organization_Assignment_List"]["Other_Positions"] = get_list_of_dicts_input(
        "Other Position",
        {
            "Name": {"type": "string", "prompt": "    Enter Name: "},
            "Position": {"type": "string", "prompt": "    Enter Position: "},
            "Agency": {"type": "string", "prompt": "    Enter Agency: "},
            "Phone": {"type": "string", "prompt": "    Enter Phone: "}
        }
    )

    # ICS Form 204 - Assignment List
    iap_data["ICS_204_Assignment_List"] = get_list_of_dicts_input(
        "Assignment",
        {
            "Division_Group": {"type": "string"},
            "Branch": {"type": "string", "prompt": "    Enter Branch (optional, press Enter to skip): ", "default": None},
            "Task_Activity": {"type": "string"},
            "Special_Instructions": {"type": "string", "prompt": "    Enter Special Instructions (optional, press Enter to skip): ", "default": None},
            "Reporting_Location": {"type": "string"},
            "Reporting_Time": {"type": "time", "format_str": "%H:%M"},
            "Supervisor_Name": {"type": "string", "prompt": "    Enter Supervisor Name: "},
            "Supervisor_Contact": {"type": "string", "prompt": "    Enter Supervisor Contact: "}
        }
    )
    # Special handling for Resources_Assigned within Assignment List
    for assignment in iap_data["ICS_204_Assignment_List"]:
        resources = []
        print(f"  --- Resources for {assignment.get('Division_Group', 'an assignment')} ---")
        while True:
            resource = get_string_input("    Enter Resource (e.g., Engine 1, Medic 5 - type 'done' to finish): ")
            if resource.lower() == 'done':
                break
            if resource:
                resources.append(resource)
        assignment["Resources_Assigned"] = resources
        # Flatting supervisor details
        assignment["Supervisor"] = {
            "Name": assignment.pop("Supervisor_Name"),
            "Contact": assignment.pop("Supervisor_Contact")
        }

    # ICS Form 205 - Incident Radio Communications Plan
    iap_data["ICS_205_Incident_Radio_Communications_Plan"] = get_list_of_dicts_input(
        "Radio Channel",
        {
            "Channel_Name": {"type": "string"},
            "Function": {"type": "string"},
            "Assignment": {"type": "string"},
            "Frequency_Mode_Tone": {"type": "string"},
            "Remarks": {"type": "string", "prompt": "    Enter Remarks (optional, press Enter to skip): ", "default": None}
        }
    )

    # ICS Form 206 - Medical Plan
    print("\n--- ICS Form 206 - Medical Plan ---")
    iap_data["ICS_206_Medical_Plan"]["Medical_Aid_Stations"] = get_list_of_dicts_input(
        "Medical Aid Station",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Capabilities": {"type": "string"}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Hospitals"] = get_list_of_dicts_input(
        "Hospital",
        {
            "Name": {"type": "string"},
            "Location": {"type": "string"},
            "Contact": {"type": "string"},
            "Specializations": {"type": "string", "prompt": "    Enter Specializations (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Ambulance_Services"] = get_list_of_dicts_input(
        "Ambulance Service",
        {
            "Provider": {"type": "string"},
            "Contact": {"type": "string"},
            "Exchange_Points": {"type": "string", "prompt": "    Enter Exchange Points (optional, press Enter to skip): ", "default": None}
        }
    )
    iap_data["ICS_206_Medical_Plan"]["Medical_Emergency_Procedures"] = get_string_input("Enter Medical Emergency Procedures: ")

    # Map/Diagram
    print("\n--- Map / Diagram ---")
    iap_data["Map_Diagram"]["Description"] = get_string_input("Enter Map/Diagram Description (optional, press Enter to skip): ", default=None)
    iap_data["Map_Diagram"]["File_Path"] = get_string_input("Enter Map/Diagram File Path/URL (optional, press Enter to skip): ", default=None)

    # Safety Message/Plan
    iap_data["Safety_Message_Plan"] = get_string_input("Enter Safety Message/Plan details (optional, press Enter to skip): ", default=None)

    return iap_data

def generate_iap_document(iap_data):
    """Generates a formatted NIMS/ICS compliant Incident Action Plan document from collected data."""
    document_parts = []

    document_parts.append("# Incident Action Plan (IAP)\n")

    if iap_data.get("Incident_Name"):
        document_parts.append(f"## Incident Name: {iap_data['Incident_Name']}\n")

    # Operational Period
    op_period = iap_data.get("Operational_Period")
    if op_period and op_period.get("Start_Date"):
        document_parts.append("### 1. Operational Period\n")
        document_parts.append(f"* **Start:** {op_period.get('Start_Date')} {op_period.get('Start_Time')}\n")
        document_parts.append(f"* **End:** {op_period.get('End_Date')} {op_period.get('End_Time')}\n\n")

    # ICS 202 - Incident Objectives
    ics_202 = iap_data.get("ICS_202_Incident_Objectives")
    if ics_202 and (ics_202.get("Objectives") or ics_202.get("General_Safety_Message")):
        document_parts.append("### 2. ICS Form 202 - Incident Objectives\n")
        if ics_202.get("Objectives"):
            document_parts.append("#### Objectives:\n")
            for obj in ics_202["Objectives"]:
                document_parts.append(f"* {obj}\n")
        if ics_202.get("General_Safety_Message"):
            document_parts.append(f"#### General Safety Message:\n{ics_202['General_Safety_Message']}\n")
        if ics_202.get("Prepared_By") and ics_202["Prepared_By"].get("Name"):
            prep = ics_202["Prepared_By"]
            document_parts.append(f"* **Prepared By:** {prep.get('Name')} ({prep.get('Position')})\n")
            if prep.get('Signature'):
                document_parts.append(f"* **Signature:** {prep.get('Signature')}\n")
            document_parts.append(f"* **Date/Time:** {ics_202.get('Date')} {ics_202.get('Time')}\n\n")

    # ICS 203 - Organization Assignment List
    ics_203 = iap_data.get("ICS_203_Organization_Assignment_List")
    if ics_203 and (ics_203.get("Incident_Commander") or ics_203.get("Section_Chiefs") or ics_203.get("Other_Positions")):
        document_parts.append("### 3. ICS Form 203 - Organization Assignment List\n")
        if ics_203.get("Incident_Commander") and ics_203["Incident_Commander"].get("Name"):
            ic = ics_203["Incident_Commander"]
            document_parts.append(f"#### Incident Commander:\n")
            document_parts.append(f"* **Name:** {ic.get('Name')}\n")
            document_parts.append(f"* **Agency:** {ic.get('Agency')}\n")
            document_parts.append(f"* **Phone:** {ic.get('Phone')}\n\n")

        if ics_203.get("Section_Chiefs"):
            document_parts.append("#### Section Chiefs:\n")
            for section, details in ics_203["Section_Chiefs"].items():
                if details.get("Name"):
                    document_parts.append(f"* **{section.replace('_', ' ')}:** {details.get('Name')} ({details.get('Agency')}) - {details.get('Phone')}\n")
            document_parts.append("\n")

        if ics_203.get("Other_Positions"):
            document_parts.append("#### Other Key Positions:\n")
            for pos in ics_203["Other_Positions"]:
                if pos.get("Name"):
                    document_parts.append(f"* **{pos.get('Position')}:** {pos.get('Name')} ({pos.get('Agency')}) - {pos.get('Phone')}\n")
            document_parts.append("\n")

    # ICS 204 - Assignment List
    ics_204_list = iap_data.get("ICS_204_Assignment_List")
    if ics_204_list:
        document_parts.append("### 4. ICS Form 204 - Assignment List\n")
        for i, assignment in enumerate(ics_204_list):
            if assignment.get("Division_Group"):
                document_parts.append(f"#### Assignment {i+1}: {assignment.get('Division_Group')}\n")
                if assignment.get("Branch"):
                    document_parts.append(f"* **Branch:** {assignment.get('Branch')}\n")
                if assignment.get("Resources_Assigned"):
                    document_parts.append(f"* **Resources Assigned:** {', '.join(assignment['Resources_Assigned'])}\n")
                if assignment.get("Task_Activity"):
                    document_parts.append(f"* **Task/Activity:** {assignment.get('Task_Activity')}\n")
                if assignment.get("Special_Instructions"):
                    document_parts.append(f"* **Special Instructions:** {assignment.get('Special_Instructions')}\n")
                if assignment.get("Reporting_Location"):
                    document_parts.append(f"* **Reporting Location:** {assignment.get('Reporting_Location')}\n")
                if assignment.get("Reporting_Time"):
                    document_parts.append(f"* **Reporting Time:** {assignment.get('Reporting_Time')}\n")
                if assignment.get("Supervisor") and assignment["Supervisor"].get("Name"):
                    sup = assignment["Supervisor"]
                    document_parts.append(f"* **Supervisor:** {sup.get('Name')} ({sup.get('Contact')})\n")
                document_parts.append("\n")

    # ICS 205 - Incident Radio Communications Plan
    ics_205_list = iap_data.get("ICS_205_Incident_Radio_Communications_Plan")
    if ics_205_list:
        document_parts.append("### 5. ICS Form 205 - Incident Radio Communications Plan\n")
        for i, channel in enumerate(ics_205_list):
            if channel.get("Channel_Name"):
                document_parts.append(f"#### Channel {i+1}: {channel.get('Channel_Name')}\n")
                document_parts.append(f"* **Function:** {channel.get('Function')}\n")
                document_parts.append(f"* **Assignment:** {channel.get('Assignment')}\n")
                document_parts.append(f"* **Frequency/Mode/Tone:** {channel.get('Frequency_Mode_Tone')}\n")
                if channel.get("Remarks"):
                    document_parts.append(f"* **Remarks:** {channel.get('Remarks')}\n")
                document_parts.append("\n")

    # ICS 206 - Medical Plan
    ics_206 = iap_data.get("ICS_206_Medical_Plan")
    if ics_206 and (ics_206.get("Medical_Aid_Stations") or ics_206.get("Hospitals") or ics_206.get("Ambulance_Services") or ics_206.get("Medical_Emergency_Procedures")):
        document_parts.append("### 6. ICS Form 206 - Medical Plan\n")
        if ics_206.get("Medical_Aid_Stations"):
            document_parts.append("#### Medical Aid Stations:\n")
            for station in ics_206["Medical_Aid_Stations"]:
                if station.get("Name"):
                    document_parts.append(f"* **{station.get('Name')}:** {station.get('Location')} (Capabilities: {station.get('Capabilities')})\n")
            document_parts.append("\n")

        if ics_206.get("Hospitals"):
            document_parts.append("#### Hospitals:\n")
            for hospital in ics_206["Hospitals"]:
                if hospital.get("Name"):
                    specs = f" (Specializations: {hospital['Specializations']})" if hospital.get('Specializations') else ''
                    document_parts.append(f"* **{hospital.get('Name')}:** {hospital.get('Location')} (Contact: {hospital.get('Contact')}){specs}\n")
            document_parts.append("\n")

        if ics_206.get("Ambulance_Services"):
            document_parts.append("#### Ambulance Services:\n")
            for service in ics_206["Ambulance_Services"]:
                if service.get("Provider"):
                    ex_points = f" (Exchange Points: {service['Exchange_Points']})" if service.get('Exchange_Points') else ''
                    document_parts.append(f"* **{service.get('Provider')}:** Contact: {service.get('Contact')}{ex_points}\n")
            document_parts.append("\n")

        if ics_206.get("Medical_Emergency_Procedures"):
            document_parts.append(f"#### Medical Emergency Procedures:\n{ics_206['Medical_Emergency_Procedures']}\n\n")

    # Map/Diagram
    map_diag = iap_data.get("Map_Diagram")
    if map_diag and (map_diag.get("Description") or map_diag.get("File_Path")):
        document_parts.append("### 7. Map / Diagram\n")
        if map_diag.get("Description"):
            document_parts.append(f"* **Description:** {map_diag.get('Description')}\n")
        if map_diag.get("File_Path"):
            document_parts.append(f"* **File Path/URL:** {map_diag.get('File_Path')}\n")
        document_parts.append("\n")

    # Safety Message/Plan
    if iap_data.get("Safety_Message_Plan"):
        document_parts.append("### 8. Safety Message / Plan\n")
        document_parts.append(f"{iap_data['Safety_Message_Plan']}\n\n")

    return "".join(document_parts)

def save_iap_to_files(formatted_iap_document, filename_prefix="incident_iap"):
    """Saves the formatted IAP document to Markdown and plain text files."""
    markdown_filename = f"{filename_prefix}.md"
    text_filename = f"{filename_prefix}.txt"

    try:
        with open(markdown_filename, "w", encoding="utf-8") as f:
            f.write(formatted_iap_document)
        print(f"IAP successfully saved to {markdown_filename}")

        # For simplicity, saving the Markdown content directly to the text file
        # A more robust solution might involve stripping markdown or converting to pure text.
        with open(text_filename, "w", encoding="utf-8") as f:
            f.write(formatted_iap_document)
        print(f"IAP successfully saved to {text_filename}")

    except IOError as e:
        print(f"Error saving files: {e}")

def save_iap_to_pdf(formatted_iap_document, filename_prefix="incident_iap"):
    """Converts the Markdown formatted IAP document to HTML and then saves it as a PDF file."""
    pdf_filename = f"{filename_prefix}.pdf"

    try:
        # Convert Markdown to HTML
        html_content = markdown.markdown(formatted_iap_document)

        # Add some basic styling for better PDF output
        html_styled = f"""
<html>
<head>
    <meta charset=\"utf-8\">
    <title>{filename_prefix.replace('_', ' ').title()} IAP</title>
    <style>
        body {{ font-family: sans-serif; margin: 1in; }}
        h1 {{ color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 0.5em; }}
        h2 {{ color: #34495e; border-bottom: 1px solid #ccc; padding-bottom: 0.3em; }}
        h3 {{ color: #34495e; margin-top: 1.5em; }}
        h4 {{ color: #34495e; margin-top: 1em; }}
        ul {{ list-style-type: disc; margin-left: 1.5em; }}
        li {{ margin-bottom: 0.5em; }}
        b {{ font-weight: bold; }}
    </style>
</head>
<body>
{html_content}
</body>
</html>
"""

        # Generate PDF from HTML
        HTML(string=html_styled).write_pdf(pdf_filename)
        print(f"IAP successfully saved to {pdf_filename}")

    except Exception as e:
        print(f"Error saving PDF file: {e}")


# --- Main Execution Flow ---
print("Starting interactive IAP data collection. Follow the prompts.")
filled_iap_data = collect_iap_data(iap_data_model)

print("\n--- Collected IAP Data ---")
print(json.dumps(filled_iap_data, indent=4))

print("\n--- Generating IAP Document ---")
formatted_iap_document = generate_iap_document(filled_iap_data)
print(formatted_iap_document)

# Save the generated document to files
save_iap_to_files(formatted_iap_document, filename_prefix="generated_iap")
save_iap_to_pdf(formatted_iap_document, filename_prefix="generated_iap")
```
